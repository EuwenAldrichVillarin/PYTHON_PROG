```python
"""
Task 7:OOP: Vehicles with Odometer and Fuel. Design classes Vehicle (base),
Car and Truck (children). Each has color, mileage, and fuel_liters. Provide
drive(km, km_per_liter) which increments mileage and decreases fuel (never below 0) and refuel(liters). Show how inheritance avoids duplication and add
a method override in Truck (e.g., different efficiency). Include a minimal test
script. Anchor concepts: classes vs. instances, attributes, instance methods, inheritance/override. Reference: :contentReference[oaicite:13]index=13

Prompt(ChatGPT):Create a python program wtih the following: A base class named Vehicle, 
Car and Truck with values of color,milage and fuel_liters. also create a drive(km,km_per_liter)
class which will increment the milage and decrease fuel (never below 0). also create a refuel 
class that increases the fuel of the car not past its fuel capacity
"""
#ChatGPT code
print("ChatGPT code")
# Base class: Vehicle
class Vehicle:
    def __init__(self, color, mileage, fuel_liters, fuel_capacity):
        self.color = color
        self.mileage = mileage
        self.fuel_liters = fuel_liters
        self.fuel_capacity = fuel_capacity

    def drive(self, km, km_per_liter):
        """Drive the vehicle for 'km' kilometers.
        Reduces fuel and increases mileage accordingly.
        Fuel will never drop below 0."""
        required_fuel = km / km_per_liter

        if required_fuel > self.fuel_liters:
            # Drive only as far as fuel allows
            possible_km = self.fuel_liters * km_per_liter
            self.mileage += possible_km
            self.fuel_liters = 0
            print(f"Ran out of fuel after driving {possible_km:.1f} km.")
        else:
            self.mileage += km
            self.fuel_liters -= required_fuel
            print(f"Drove {km} km. Fuel left: {self.fuel_liters:.1f} L.")

    def refuel(self, liters):
        """Add fuel to the vehicle but do not exceed capacity."""
        if self.fuel_liters + liters > self.fuel_capacity:
            added = self.fuel_capacity - self.fuel_liters
            self.fuel_liters = self.fuel_capacity
            print(f"Refueled {added:.1f} L to reach full capacity ({self.fuel_capacity} L).")
        else:
            self.fuel_liters += liters
            print(f"Refueled {liters:.1f} L. Current fuel: {self.fuel_liters:.1f} L.")

    def display_info(self):
        print(f"{self.__class__.__name__} Info ‚Üí Color: {self.color}, "
              f"Mileage: {self.mileage:.1f} km, Fuel: {self.fuel_liters:.1f} L")


# Derived class: Car
class Car(Vehicle):
    def __init__(self, color, mileage, fuel_liters):
        super().__init__(color, mileage, fuel_liters, fuel_capacity=50)  # Example capacity: 50 L


# Derived class: Truck
class Truck(Vehicle):
    def __init__(self, color, mileage, fuel_liters):
        super().__init__(color, mileage, fuel_liters, fuel_capacity=120)  # Example capacity: 120 L


# Example usage
if __name__ == "__main__":
    car = Car("Red", 20000, 10)
    truck = Truck("Blue", 50000, 50)

    car.display_info()
    car.drive(100, 15)   # 15 km per liter efficiency
    car.refuel(30)
    car.display_info()

    print("-" * 40)

    truck.display_info()
    truck.drive(600, 8)  # 8 km per liter efficiency
    truck.refuel(100)
    truck.display_info()
_=="""
Critique
    Correctness: The program correctly implements inheritance and method overriding to model vehicles
    , cars, and trucks. The drive() and refuel() methods logically update mileage and fuel levels 
    while preventing values from exceeding capacity or dropping below zero. However, it lacks user
    input to make it interactive and error handling for invalid values (e.g., negative distances or over-refueling).

¬†¬†¬†¬†Time and Space Complexity: Both are O(1) since operations such as fuel updates and arithmetic 
    calculations occur in constant time and space, regardless of input size.

¬†¬†¬†¬†Robustness: The code runs correctly with preset values but could be improved by adding 
    user input prompts and validation checks. It should restrict inputs that exceed the car‚Äôs 
    fuel capacity or attempt to drive beyond feasible limits. Exception handling (e.g., ValueError) 
    would make it more fault-tolerant.

¬†¬†¬†¬†Readability: The structure is clear, with descriptive method names, consistent formatting, 
    and informative print statements. The class hierarchy and use of super() enhance organization 
    and understanding.

¬†¬†¬†¬†Faithfulness: It follows Lesson 3 (Function Creation),
    Lesson 4 (Conditionals), 
    Lesson 6 (Object-Oriented Programming and Inheritance),
    and partially Lesson 1 (User Input)
"""
print()
print("Improved Code")
class Vehicle:
    def __init__(self, color, mileage, fuel_liters, fuel_capacity):
        self.color = color
        self.mileage = mileage
        self.fuel_liters = fuel_liters
        self.fuel_capacity = fuel_capacity

    def drive(self, km, km_per_liter):
        """Drive the vehicle for 'km' kilometers with given efficiency."""
        if km <= 0 or km_per_liter <= 0:
            print("Distance and efficiency must be positive values.")
            return

        required_fuel = km / km_per_liter

        if required_fuel > self.fuel_liters:
            possible_km = self.fuel_liters * km_per_liter
            self.mileage += possible_km
            self.fuel_liters = 0
            print(f"Not enough fuel! You only drove {possible_km:.1f} km before running out.")
        else:
            self.mileage += km
            self.fuel_liters -= required_fuel
            print(f"Drove {km:.1f} km. Remaining fuel: {self.fuel_liters:.1f} L.")

    def refuel(self, liters):
        """Add fuel up to capacity."""
        if liters <= 0:
            print("Refuel amount must be positive.")
            return

        if self.fuel_liters >= self.fuel_capacity:
            print("Tank is already full.")
            return

        if self.fuel_liters + liters > self.fuel_capacity:
            added = self.fuel_capacity - self.fuel_liters
            self.fuel_liters = self.fuel_capacity
            print(f" Only {added:.1f} L added to reach full capacity ({self.fuel_capacity} L).")
        else:
            self.fuel_liters += liters
            print(f"Added {liters:.1f} L. Current fuel: {self.fuel_liters:.1f} L.")

    def display_info(self):
        print(f"\n {self.__class__.__name__} Info")
        print(f"Color       : {self.color}")
        print(f"Mileage     : {self.mileage:.1f} km")
        print(f"Fuel Level  : {self.fuel_liters:.1f} L / {self.fuel_capacity} L\n")


class Car(Vehicle):
    def __init__(self, color, mileage, fuel_liters):
        super().__init__(color, mileage, fuel_liters, fuel_capacity=50)


class Truck(Vehicle):
    def __init__(self, color, mileage, fuel_liters):
        super().__init__(color, mileage, fuel_liters, fuel_capacity=120)


def get_float_input(prompt):
    """Helper to ensure numeric-only float input."""
    while True:
        value = input(prompt).strip()
        try:
            # Reject empty or non-numeric values
            if not value or value.isalpha():
                raise ValueError
            number = float(value)
            return number
        except ValueError:
            print("Invalid input! Please enter a valid numeric value.")


def main():
    print("Welcome to the Vehicle Simulator \n")

    # Choose vehicle type
    while True:
        choice = input("Choose your vehicle (car/truck): ").strip().lower()
        if choice in ["car", "truck"]:
            break
        else:
            print("Invalid choice. Please type 'car' or 'truck'.")

    color = input("Enter vehicle color: ").strip()
    mileage = 0
    fuel_liters = 0

    vehicle = Car(color, mileage, fuel_liters) if choice == "car" else Truck(color, mileage, fuel_liters)
    print(f"\nCreated a {color.capitalize()} {choice.capitalize()} with empty tank.\n")

    while True:
        print("Actions: [1] Drive  [2] Refuel  [3] Show Info  [4] Exit")
        action = input("Enter your choice: ").strip()

        if action == "1":
            km = get_float_input("Enter kilometers to drive: ")
            efficiency = get_float_input("Enter km per liter: ")
            vehicle.drive(km, efficiency)

        elif action == "2":
            liters = get_float_input("Enter liters to refuel: ")
            vehicle.refuel(liters)

        elif action == "3":
            vehicle.display_info()

        elif action == "4":
            print("Bye")
            break

        else:
            print("Invalid choice. Try again.")

        print("-" * 50)


if __name__ == "__main__":
    main()
__=="""
ChatGPT said:
Discussion: Inheritance avoids duplication because both Car and Truck share the same
properties and methods (like drive, refuel, and display_info) from the Vehicle class 
instead of rewriting them. This keeps the code shorter and easier to maintain. For example,
the Truck class can override the drive method to have a different fuel efficiency, showing how 
inheritance allows customizing behavior without changing the base class.
"""
```

```python
"""
Task 8:OOP + Composition: Kennel with Dogs. Using a Dog class (name, age, breed,
bark()), implement a Kennel manager that holds multiple Dog instances, supports
add_dog, remove_dog, oldest_dog(), and find_by_breed(breed). Compare composition vs. inheritance for this scenario. Anchor concepts: OOP design, managing
collections of objects. Reference: :contentReference[oaicite:14]index=14

Prompt(ChatGPT): Create a python program that has a class named Dog(name, age, breed, bark()). 
Create a dog manager that can hold mulitple instances of a dog and also have a functionality
to add_dog,remove_dog,oldest_dog(), and find_dog_by_breed(breed)
"""
#ChatGPT code
print("ChatGPT code")
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self):
        return f"{self.name} says: Woof woof!"


class DogManager:
    def __init__(self):
        self.dogs = []

    def add_dog(self, dog):
        """Adds a Dog instance to the manager."""
        self.dogs.append(dog)
        print(f"{dog.name} has been added.")

    def remove_dog(self, name):
        """Removes a dog by its name."""
        for dog in self.dogs:
            if dog.name.lower() == name.lower():
                self.dogs.remove(dog)
                print(f"{dog.name} has been removed.")
                return
        print(f"No dog found with the name '{name}'.")

    def oldest_dog(self):
        """Returns the oldest dog."""
        if not self.dogs:
            print("No dogs in the manager.")
            return None
        oldest = max(self.dogs, key=lambda d: d.age)
        print(f"The oldest dog is {oldest.name}, age {oldest.age}.")
        return oldest

    def find_dog_by_breed(self, breed):
        """Finds and returns all dogs of a specific breed."""
        matched_dogs = [dog for dog in self.dogs if dog.breed.lower() == breed.lower()]
        if matched_dogs:
            print(f"Dogs of breed '{breed}': {[dog.name for dog in matched_dogs]}")
        else:
            print(f"No dogs of breed '{breed}' found.")
        return matched_dogs


# Example usage
if __name__ == "__main__":
    manager = DogManager()

    dog1 = Dog("Buddy", 5, "Golden Retriever")
    dog2 = Dog("Max", 3, "Bulldog")
    dog3 = Dog("Bella", 7, "Golden Retriever")

    manager.add_dog(dog1)
    manager.add_dog(dog2)
    manager.add_dog(dog3)

    print(dog1.bark())  # show bark output
    manager.oldest_dog()
    manager.find_dog_by_breed("Golden Retriever")
    manager.remove_dog("Max")

_=="""
Critique
    Correctness: The program accurately demonstrates object-oriented principles by defining Dog and DogManager 
    classes. Each method performs its intended function‚Äîadding, removing, searching, and identifying the oldest
    dog‚Äîcorrectly and logically. However, it lacks user input for dynamic interaction and does not handle invalid 
    or duplicate entries, which could cause inconsistencies in the list.

¬†¬†¬†¬†Time and Space Complexity: Adding or removing dogs has O(n) complexity in the worst case since 
    it may search through the list, while space complexity is O(n) due to storing all dog instances.

¬†¬†¬†¬†Robustness: The program works as expected for hardcoded examples but could benefit from user 
    input and input validation. It should include error handling for empty lists, duplicate dog names, 
    and invalid age or breed entries to prevent logic errors or crashes during runtime.

¬†¬†¬†¬†Readability: The structure is clear, with well-defined classes, concise method names, and docstrings 
    that explain each function. The print statements give useful feedback, and naming conventions make the
    program easy to follow and understand.

¬†¬†¬†¬†Faithfulness: It follows Lesson 3 (Function Creation),
    Lesson 4 (Conditionals),
    Lesson 5 (Use of Lists for Storage), 
    and Lesson 6 (Object-Oriented Programming ‚Äî Classes, Objects, and Methods),
"""
print()
print("Improved Code")
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self):
        return f"{self.name} says: Woof woof!"


class DogManager:
    def __init__(self):
        self.dogs = []

    def add_dog(self, dog):
        """Adds a Dog instance to the manager."""
        self.dogs.append(dog)
        print(f"{dog.name} has been added.")
        print(dog.bark())  # Make the dog bark when added!

    def remove_dog(self, name):
        """Removes a dog by its name."""
        for dog in self.dogs:
            if dog.name.lower() == name.lower():
                self.dogs.remove(dog)
                print(f"{dog.name} has been removed.")
                return
        print(f"No dog found with the name '{name}'.")

    def oldest_dog(self):
        """Returns the oldest dog."""
        if not self.dogs:
            print("No dogs in the manager.")
            return None
        oldest = max(self.dogs, key=lambda d: d.age)
        print(f"The oldest dog is {oldest.name}, age {oldest.age}.")
        return oldest

    def find_dog_by_breed(self, breed):
        """Finds and returns all dogs of a specific breed."""
        matched_dogs = [dog for dog in self.dogs if dog.breed.lower() == breed.lower()]
        if matched_dogs:
            print(f"Dogs of breed '{breed}': {[dog.name for dog in matched_dogs]}")
        else:
            print(f"No dogs of breed '{breed}' found.")
        return matched_dogs

    def make_dog_bark(self, name):
        """Finds a dog by name and makes it bark."""
        for dog in self.dogs:
            if dog.name.lower() == name.lower():
                print(dog.bark())
                return
        print(f"No dog found with the name '{name}'.")


def get_dog_input():
    """Safely get user input for dog creation."""
    while True:
        name = input("Enter dog's name: ").strip()
        if not name:
            print("Name cannot be empty.")
            continue

        try:
            age = int(input("Enter dog's age: "))
            if age < 1:
                print("Age must be at least 1.")
                continue
        except ValueError:
            print("Age must be a valid number.")
            continue

        breed = input("Enter dog's breed: ").strip()
        if not breed:
            print("Breed cannot be empty.")
            continue

        return Dog(name, age, breed)


def main():
    manager = DogManager()

    while True:
        print("1. Add a dog")
        print("2. Remove a dog")
        print("3. Show oldest dog")
        print("4. Find dog by breed")
        print("5. List all dogs")
        print("6. Make a dog bark")
        print("7. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            dog = get_dog_input()
            manager.add_dog(dog)

        elif choice == "2":
            name = input("Enter the dog's name to remove: ").strip()
            manager.remove_dog(name)

        elif choice == "3":
            manager.oldest_dog()

        elif choice == "4":
            breed = input("Enter breed to search: ").strip()
            manager.find_dog_by_breed(breed)

        elif choice == "5":
            if not manager.dogs:
                print("No dogs added yet.")
            else:
                print("Current dogs:")
                for d in manager.dogs:
                    print(f" - {d.name}, Age {d.age}, Breed: {d.breed}")

        elif choice == "6":
            name = input("Enter the dog's name to bark: ").strip()
            manager.make_dog_bark(name)

        elif choice == "7":
            print("Exiting program. Goodbye!")
            break

        else:
            print("Invalid option. Please try again.")


if __name__ == "__main__":
    main()
__=="""
Discussion: In this program, composition is used instead of inheritance because the DogManager 
class has Dog objects rather than being a Dog. Composition makes sense here since the manager‚Äôs 
job is to organize and handle multiple dogs, not to act like one. If inheritance were used, it would
not fit the relationship (a manager is not a dog). Composition keeps the classes independent and reusable‚ÄîDog
handles its own behavior like barking, while DogManager focuses on managing a collection of dogs without
duplicating code or forcing an unnatural parent-child relationship.
"""
```
